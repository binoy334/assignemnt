```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Load the dataset
df = pd.read_csv(r"C:\Users\binoy\OneDrive\Desktop\entri calss\assignment\house_price.csv")
#Identify numerical columess
numerical_columns = df.select_dtypes(include = [np.number]).columns.tolist()
#Calculate correlations
correlation_matrix = df[numerical_columns].corr()
#Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap ='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
plt.figure(figure = (10, 5))
sns.boxplot (df['price_per_sqft'])
plt.title('box plot of price per square foot')
plt.show
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(df['price_per_sqft'], kde= True)
plt.title('distribution of price per square foot')
plt.show()
```

```python
Q1 = df['price_per_sqft'].quantile(0.25)
Q3 = df['price_per_sqft'].quantile(0.50)
IQR = Q3 - Q1
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 +(1.5*IQR)
outliers_iqr = df[(df['price_per_sqft'] >upper_bound) | (df['price_per_sqft']< lower_bound)]
df_cleaned_iqr = df.drop(outliers_iqr.index).reset_index(drop=True)
```
